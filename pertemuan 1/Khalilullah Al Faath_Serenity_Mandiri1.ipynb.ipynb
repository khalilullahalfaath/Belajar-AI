# Korelasi antar tiap variabel

```python
sns.pairplot(data, hue="Species")
```

```python
fig, axs = plt.subplots(ncols=4)
sns.set(rc={'figure.figsize':(30,10)})
sns.distplot(data['SepalLengthCm'], ax=axs[0])
sns.distplot(data['SepalWidthCm'], ax=axs[1])
sns.distplot(data['PetalLengthCm'], ax=axs[2])
sns.distplot(data['PetalWidthCm'], ax=axs[3])
```

Density plot di atas merepresentasikan kepadatan data di angka tertentu. Dapat dilihat bahwa:


*   SepalLengthCm terpusat di panjang sekitar 5 cm
*   SepalWidthCm terpusat di lebar sekitar 3 cm
*   PetalLengthCm terpusat di panjang sekitar 2 cm dan sekitar 5 cm
*   PetalwidthCm terpusat di lebar sekitar 0.5 cm





```python
sns.set()
plt.figure(figsize=(10,5))
sns.countplot(data['Species'])
plt.xlabel('Jumlah individu')
plt.title('Jumlah individu berdasarkan spesies', fontweight='bold')
plt.show()
```

Dapat dilihat bahwa pada dataset jumlah tiap spesies seimbang

# Korelasi antar variabel

```python
fig, ax = plt.subplots(figsize=(10, 8))

mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))

heatmap = sns.heatmap(data.corr(), mask=mask, vmin=-1, vmax=1, annot=True)

plt.suptitle("Correlation Plot", ha='left', x=0.155, y=1.04, fontsize=18, fontweight='bold')
plt.title("Korelasi antara beberapa fitur", loc='left', fontsize=12) 
plt.tight_layout()
plt.show()
```

Dapat dilihat bahwa SepalLenghtCm berkorelasi positif terbesar dengan PetalLength dan PetalLength dan Sepal width memiliki korelasi negatif terkecil
